```
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import numpy as np

from keras.models import model_from_json
from keras.models import load_model
from matplotlib import pyplot as pyplot
import cv2
import random
import pickle

DATADIR = "train"
file_list = []
class_list = []

CATOGRIES = ["Bread","Dairy product","Dessert","Egg","Fried food","Meat","Noodles-Pasta","Rice","Seafood","Soup","Vegetable-Fruit"]

IMG_SIZE = 50
training_data = []

def create_training_Data():
    for CAT in CATOGRIES :
        count = 1
        path = DATADIR +"/"+ CAT
        class_num = CATOGRIES.index(CAT)
        for img in os.listdir(path):
            try :
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array,(IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, class_num])
                print(str(count)+"="+str(CAT))
                count = count+1
            except Exception as e:
                print("Wrong")
                pass

create_training_Data()

random.shuffle(training_data)

X = []
y = []

for features, label in training_data:
	X.append(features)
	y.append(label)


X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1)

#Creating the files containing all the information about your model
pickle_out = open("X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()

# Made the files

pickle_in = open("X.pickle", "rb")
X = pickle.load(pickle_in)
```

```
X = pickle.load(open("X.pickle", "rb"))
y = pickle.load(open("y.pickle", "rb"))

X = np.array(X)
y = np.array(y)

X = X/255.0

#3 Convolutional layers
model = Sequential() #Creation of a model
model.add(Conv2D(32, (3, 3), input_shape = X.shape[1:]))
model.add(Activation("relu")) #Rectified lu Activation
model.add(MaxPooling2D(pool_size=(2,2)))  #Downsamples to a 2X2 pool IMG_SIZE


model.add(Conv2D(64, (3, 3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

# Adding few hidden ones
model.add(Flatten())
model.add(Dense(128))
model.add(Activation("relu"))

model.add(Dense(11))
model.add(Activation("relu"))

model.add(Dense(11))
model.add(Activation("softmax"))

model.compile(loss="sparse_categorical_crossentropy",
				optimizer="adam",
				metrics=["accuracy"])

history = model.fit(X, y, batch_size=32, epochs=50, validation_split=0.1)

model_json = model.to_json()
with open("model.json", "w") as json_file :
	json_file.write(model_json)

model.save_weights("model.h5")
print("Saved model to disk")

model.save('CNN.model')
```

```
import os
import cv2
import csv
import tensorflow as tf

CATOGRIES = ["Bread","Dairy product","Dessert","Egg","Fried food","Meat","Noodles-Pasta","Rice","Seafood","Soup","Vegetable-Fruit"]
path = "test"

def prepare(file):
    IMG_SIZE = 50
    img_array = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)

def writecsv(img,output):
    with open('predict_v0.csv', 'a+', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([img,output])

model = tf.keras.models.load_model("CNN.model")
for img in os.listdir(path):
    image = prepare("test/"+img)
    prediction = model.predict([image])
    prediction = list(prediction[0])
    output = (CATOGRIES[prediction.index(max(prediction))])
    print("Prediction for "+path+ " is : " + output + img)
    writecsv(img,output)
```
